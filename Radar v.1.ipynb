```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# http://www.b3.com.br/pt_br/market-data-e-indices/servicos-de-dados/market-data/consultas/
#boletim-diario/arquivos-para-download/
df_ti = pd.read_csv('TradeInformationConsolidatedFile_20200626_1.csv')
print('Numero de códigos na base de dados= ',len(df_ti))

# Lista que irá conter os códigos filtrados
list_tickers = []

# Iteração da lista de ativos
for i in df_ti.index:
    #TradQty = Quantidade de negócios no dia maior que 1000
    # CASH = Mercado à vista de ações
    if (df_ti['TradQty'][i] > 1000) and (df_ti['SgmtNm'][i] == 'CASH'):
        list_tickers.append(df_ti['TckrSymb'][i]+'.SA')

print('Numero de códigos filtrados = ',len(list_tickers)) 
```

```python
# Yahoo Finance: download dos dados dos ativos com os códigos na lista list_tickers
df = yf.download(list_tickers,group_by="ticker", start="2020-01-01", end="2030-01-01")
```

```python
#cria uma variável de indexação do Pandas
idx = pd.IndexSlice
#cria novo DF contendo somente as colunas 'Adj Close'
df_ac = df.loc[:,idx[:,'Adj Close']]
#retira o primeiro nível da multiindexação (Adj. Close) para ficar apenas com o código do ativo
df_ac.columns = df_ac.columns.droplevel(1)
#ordena as columas do DF por nome
df_ac = df_ac.sort_index(axis=1)
#Bollinger Bands(BB): cria DF dos ativos cotendo a relação [(preço - média móvel (20d))/DP(20d)]
df_desvio = pd.DataFrame()
#cálculo do DF com o valor do BB
for c in df_ac:      
    desvio = (df_ac[c]-df_ac[c].rolling(20).mean())/df_ac[c].rolling(20).std()
    df_desvio = pd.concat([df_desvio , desvio] , axis=1)
    
df_desvio
```

```python
#cria a lista com o índice dos ativos para armazenar os valores dos últimos valores das BB
ind = df_desvio.columns
ind = ind.tolist()
#cria a lista com os valores dos ativos
valores = df_desvio.iloc[-1].tolist()
#cria a lista com os ativos e os valores
desvio = pd.Series(valores, index = ind)
```

```python
print('10 ativos mais desvalorizados')
print(desvio.sort_values(ascending=True).head(10))
print()
print('10 ativos mais valorizados')
print(desvio.sort_values(ascending=False).head(10))
```
