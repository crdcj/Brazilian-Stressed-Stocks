```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import investpy as inv
import time

# http://www.b3.com.br/pt_br/market-data-e-indices/servicos-de-dados/market-data/consultas/
#boletim-diario/arquivos-para-download/
df_ti = pd.read_csv('TradeInformationConsolidatedFile_20200626_1.csv')
print('Numero de códigos na base de dados= ',len(df_ti))
n_tickers = 0

# Create a list of active tickers
for i in range(len(df_ti)):
    if df_ti.iloc[i,-2]    > 1000:
        if df_ti.iloc[i,3] == 'CASH':
            #print(df_ti.iloc[i,1])
            if n_tickers == 0:
                list_tickers = pd.Series(df_ti.iloc[i,1]+'.SA')
                n_tickers += 1
            else:                
                list_tickers = pd.concat([list_tickers, pd.Series([df_ti.iloc[i,1]+'.SA'])],ignore_index=True)
                n_tickers += 1
    
list_tickers = list_tickers.tolist() 
print('Number of Tickers = ',len(list_tickers)) 
```

```python
df = yf.download(list_tickers,group_by="ticker", start="2017-01-01", end="2030-01-01")
df
```

```python
idx = pd.IndexSlice
columns =df.loc[:,idx[:,'Adj Close']]
df1 = df
for c in columns:
    df1[c[0],'X_Desvio']  = (df[c]-df[c].rolling(20).mean())/df[c].rolling(20).std()
     
df1 = df1.sort_index(axis=1)
df1
```

```python
df_desvio =df1.loc[:,idx[:,'X_Desvio']]
df_desvio 
```

```python
ind = df_desvio.columns.levels[0]
ind = ind.tolist()
valores = df_desvio.iloc[-1].tolist()
desvio = pd.Series(valores, index = ind)

desvio
```
